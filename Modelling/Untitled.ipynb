```python
#SUPPORTING FUNCTIONS
def find_B(h, w, alpha):
    a = np.radians(alpha)
    ca = np.cos(a)
    sa = np.sin(a)
    B = np.arccos((h*ca**2-h*sa**2+h+2*w*sa*ca)/(np.sqrt(h**2*sa**4-2*h**2*sa**2+h**2*ca**4+2*h**2*ca**2+2*h**2*sa**2*ca**2+h**2+4*h*w*sa*ca+w**2*sa**4+w**2*ca**4+2*w**2*sa**2*ca**2)))
    return(np.degrees(B))

def find_kappa(alpha, betta):
    return(2*alpha + betta)

def find_koef(angle, mapping, accuracy = 0):
    a = round(angle, accuracy)
    if a in mapping.keys():
        return(mapping[a])
    else:
        return(0)
    
def print_angles(alp, bet, kap):
    for a,b,k in zip(alp,bet,kap):
        print("alpha: %.2f \t betta: %.2f \t kappa: %.2f" % (a,b,k))
```

```python
#CALCULATIONS
for w in widths:
    for h in heights:
        #calculate angles
        bettas = [find_B(h, w, a) for a in alphas]
        kappas = [find_kappa(a, b) for a,b in zip(alphas, bettas)]
        #calculate energy
        energy_emitted = [energy * find_koef(b, betta_map) for b in bettas]        
        energy_received = [e * find_koef(k, kappa_map) for k, e in zip(kappas, energy_emitted)]
        print("="*50)
        print("height = %d mm, width = %d mm" % (h,w))
        print("="*50)
        plot.plot(alphas, energy_received)
        plot.xlabel("Angle")
        plot.ylabel("Energy")
        plot.show()
        #print_angles(alphas, bettas, kappas)
    print("-"*20+"WIDTH CHANGE"+"-"*20)
```
